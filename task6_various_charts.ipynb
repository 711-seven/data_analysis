```python
import tushare as ts
from pyecharts.charts import Liquid, Gauge, WordCloud, Line, Grid
from pyecharts import options as opts
```

```python
# 导入 tushare 数据包，提取 A 股基本数据信息，取毛利率 gpr 和净利率 npr 的平均数，并分别以水球图 Liquid 和仪表盘图 Gauge 表现出来

basics = ts.get_stock_basics()
gpr_mean = round(basics["gpr"].mean(),2)/100
npr_mean = round(basics["npr"].mean(),2)

l = (
    Liquid(init_opts=opts.InitOpts(width="500px", height="500px"))
    .add("", [gpr_mean, gpr_mean])
    .set_global_opts(title_opts=opts.TitleOpts(title="A股平均毛利率"))
    )


l.render_notebook()
```

```python
g = (
    Gauge(init_opts=opts.InitOpts(width="500px", height="500px"))
    .add("",[("A股平均净利率", npr_mean)],radius="60%")
    .set_global_opts(title_opts=opts.TitleOpts(title="A股基本数据信息"))
    )

g.render_notebook()
```

```python
# 统计 A 股各行业公司数量，并跟进行业分布数据作词云图

basics.head()
```

```python
df_gb = basics.groupby("industry")["name"].agg([("number_of_comp","count")])
industry = df_gb.index.tolist()
number = df_gb.values.reshape(-1).tolist()
word = [tuple(z) for z in zip(industry, number)] 
word
word_cloud = (
    WordCloud()
    .add("", word)
    .set_global_opts(title_opts=opts.TitleOpts(title="A股各行业公司分布"))
)
word_cloud.render_notebook()
```

```python
# 利用 tushare 提取茅台数据中的最高最低价，作双折线图，茅台股票代码 600519

df = ts.get_hist_data("600519")

attr = df.index.tolist()[::-1]
high = df["high"].tolist()[::-1]
low = df["low"].tolist()[::-1]

lines = (
    Line(init_opts=opts.InitOpts(width="900px",height="500px" ))
    .add_xaxis(attr)
    .add_yaxis("high", high, label_opts=opts.LabelOpts(position="top"))
    .add_yaxis("low", low, label_opts=opts.LabelOpts(position="bottom"))
    .set_global_opts(title_opts=opts.TitleOpts(title="茅台股价high-low趋势"),
                     yaxis_opts=opts.AxisOpts(min_=400))
    )
lines.render_notebook()
```
