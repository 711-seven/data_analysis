```python
from pyecharts.charts import Bar, Pie   # 导入相关图形的构造方法
from pyecharts import options as opts   # 引入配置项入口
import pandas as pd

from warnings import simplefilter
simplefilter(action="ignore", category=FutureWarning)
```

```python
with open("C:/Users/seven/Documents/Python Scripts/Task_4_lianjia_sale.csv", "r", encoding="utf-8") as f:
    df = pd.read_csv(f)
    display(df.head()) 
```

```python
# 以【区域】字段进行 groupby， 同时结合 agg 聚合函数统计各个区域平均房价;
# 将统计所得平均房价命名为【区域平均价格】列
# 并将数据从小到大按顺序排列

# df_gb = df.groupby("区域")["价格"].agg([('区域平均价格', 'mean'),('区域最高价格', 'max'), ...])   --用tup代替dict 新写法

df_gb = df.groupby("区域")["价格"].agg({"区域平均价格":"mean"})
df_avg = df_gb.sort_values(by = "区域平均价格", ascending = True)
df_avg.head()
```

```python
# 利用 Pyecharts 作条形图，图表标题为“上海各区域平均价格”

bar = (
    Bar()
    .add_xaxis(df_avg.index.tolist())     # 需要转化为list
    .add_yaxis("区域平均价格",df_avg.values.round(decimals=1).reshape(-1).tolist())   # 需要转化为list 否则显示不出来
    .set_global_opts(title_opts=opts.TitleOpts(title="上海各区域平均价格"))
                    )
bar.render_notebook()
```

```python
# 以【装修类型】字段进行 groupby，同时结合 agg 聚合函数统计各个区域房源数量

df_zhuangxiu = df.groupby("装修类型")["房型"].agg([("房源数量", "count")])
df_zhuangxiu
```

```python
# 利用 Pyecharts 将第(3)问所得数据作饼图(阅读参考文章 7、8)，图表标题为“装修类型占比情况”
# (提示: v1=["啤酒","可乐","雪碧","咖啡","奶茶"] v2=[30,19,21,12,18]，a = [list(z) for z in zip(v1,v2)]

v1 = df_zhuangxiu.index.tolist()
v2 = df_zhuangxiu.values.reshape(-1).tolist()
pair = [list(z) for z in zip(v1, v2)]

pie = (
    Pie()
    .add("",pair,radius="50%")
    .set_global_opts(title_opts=opts.TitleOpts(title="装修类型占比情况"))
    .set_series_opts(label_opts=opts.LabelOpts(formatter="{b}:{c}, {d}%"))
    )

pie.render_notebook()
```
