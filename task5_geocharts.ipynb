```python
import pandas as pd
import tushare as ts
from pyecharts.charts import Map, Geo
from pyecharts import options as opts
```

```python
with open ("C:/Users/seven/Documents/Python Scripts/Task_4_lianjia_sale.csv","r",encoding="utf-8") as f:
    df = pd.read_csv(f)
display(df.head())
```

```python
# 根据上海各个区的房均价数据，利用 Pyecharts 中的 Map，以上海地图为背景作可视化展示图

df_gb = df.groupby("区域")["价格"].agg([("房屋均价", "mean")])
district = ['嘉定区','奉贤区','宝山区','徐汇区','普陀区','杨浦区','松江区','浦东新区','虹口区','金山区',
 '长宁区','闵行区','青浦区','静安区','黄浦区']
price_avg = df_gb.values.reshape(-1).round(decimals=2).tolist()
lst = [list(z) for z in zip(district, price_avg)]

c = (
    Map()
    .add("上海市各区县房屋均价", lst, maptype="上海")
    .set_global_opts(visualmap_opts=opts.VisualMapOpts(is_show=True, min_=min(price_avg), max_=max(price_avg)),
                    title_opts=opts.TitleOpts(title="上海市各区县房屋均价"))
    )
c.render_notebook()
```

```python
# 导入 tushare 数据包，提取 A 股基本数据信息，利用 groupby 函数统计每个省份的公司数量，并以中国地图 为背景作相应图表

df_comp = ts.get_stock_basics()
df_comp.head()
```

```python
df_comp_gb = df_comp.groupby("area")["name"].agg([("company_count","count")])
cities = df_comp_gb.index.tolist()
num = df_comp_gb.values.reshape(-1).tolist()
lst2 = [list(z) for z in zip(cities, num)]

entities = (
    Map()
    .add("中国各省上市公司数量", lst2, maptype="china")
    .set_global_opts(visualmap_opts=opts.VisualMapOpts(is_show=True, min_=min(num), max_=max(num)),
                     title_opts=opts.TitleOpts(title="中国各省上市公式数量"))
            )

entities.render_notebook()
                     
    
```
